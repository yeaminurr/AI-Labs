```python
import pandas as pd
import numpy as num
mydata = pd.read_csv('glass source classification dataset.csv')
mydata.head()
```

```python
import pandas as pd
mydata.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=num.nan, strategy='mean')
impute.fit(mydata[['Ca']])
mydata['Ca']=impute.transform(mydata[['Ca']])
mydata
```

```python
mydata.Ba = [1 if i == "Does not exist" else 0 for i in mydata.Ba]
mydata.Fe = [1 if i == "Does not exist" else 0 for i in mydata.Fe]
mydata['Type'] = mydata['Type'].map({'building_window glass':1,'vehicle_window glass':0,'container glass':2,'headlamp glass':3,'tableware glass':4})
mydata.head()
#mydata.to_csv('Lab5v2.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler
myscale = MinMaxScaler()
myscale.fit(mydata)
scaledset = myscale.transform(mydata)
mydata = pd.DataFrame(scaledset,columns=[' ','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type'])
mydata
```

```python
from sklearn.model_selection import train_test_split
newdata=mydata.drop(['Type'], axis = 1)
g=mydata['Type']
X_train, X_test, y_train, y_test = train_test_split(newdata, g, test_size=0.33)
y_test.shape
X_train.shape
mydata.shape
```
